```python
#Creation of the item list 

listo = []
for page in orders:
    for i in page:
        try:
            testo = i['line_items'][0]['parent_name']
            testo2 = testo.split("-")
            listo.append(testo2)
            item_list = [i[0] for i in listo]
        except AttributeError:
                item_list.append('Null') 
                

print(item_list)
```
